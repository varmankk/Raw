```python
data = pd.read_csv('/kaggle/input/microsoft-adventure-works-cycles-customer-data/AWCustomers.csv')
data
```

```python
data.info()

data.isnull().sum()
```

```python
data.shape
```

```python
data.drop(['Title','MiddleName','Suffix','AddressLine2'],axis=1,inplace=True)
data.info()
```

```python
data.drop(['CustomerID','FirstName','LastName','City','StateProvinceName','PhoneNumber','Education','HomeOwnerFlag','LastUpdated'],axis=1,inplace=True)
data.info()
```

```python
data.drop(['AddressLine1','CountryRegionName'],axis=1,inplace=True)
data.drop('Occupation',axis=1,inplace=True)
data.info()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
data['YearlyIncome_Normalised']=scaler.fit_transform(data['YearlyIncome'].values.reshape(-1,1))
data['YearlyIncome_Normalised']
```

```python
binned_yearly_income=np.arange(min(data['YearlyIncome']),max(data['YearlyIncome']),20000)
len(binned_yearly_income)
```

```python
YearlyIncome_label=['below low','low','Medium','above medium','High']
data['YearlyBinned']=pd.cut(data['YearlyIncome'],bins=binned_yearly_income,labels=YearlyIncome_label)
data['YearlyBinned']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data['YearlyIncome_Normalised_standard']=sc.fit_transform(data['YearlyIncome'].values.reshape(-1,1))
data
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
onehot=ohe.fit_transform(data['Gender'].values.reshape(-1,1))
ohe.categories_
```

```python
onehot.toarray()
```

```python
ohe.get_feature_names_out(['Gender'])
```

```python
gender_encoded_data=pd.DataFrame(onehot.toarray(),columns=ohe.get_feature_names_out(['Gender']))
gender_encoded_data
```

```python
data=pd.concat([data.drop('Gender',axis=1),gender_encoded_data],axis=1)
data.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder_income=OrdinalEncoder()
data['YearlyBinned']=ordinal_encoder_income.fit_transform(data['YearlyBinned'].values.reshape(-1,1))
data
```

```python
data.isnull().sum()
```

```python
data['YearlyBinned'].mean
```

```python
data.fillna(data['YearlyBinned'].mean,axis=1,inplace=True)
data['YearlyBinned'].value_counts()
```

```python
data['Year'] = data['BirthDate'].apply(lambda x: int(x.split('-')[0]))
data
```

```python
data.drop('BirthDate',axis=1,inplace=True)
data['MaritalStatus'].value_counts()
```

```python
onehot=ohe.fit_transform(data['MaritalStatus'].values.reshape(-1,1))
ohe.categories_
```

```python
marriage_encoded_data=pd.DataFrame(onehot.toarray(),columns=ohe.get_feature_names_out(['MaritalStatus']))
data=pd.concat([data.drop('MaritalStatus',axis=1),marriage_encoded_data],axis=1)
data.head()
```

```python
data.drop('PostalCode',axis=1,inplace=True)
data.head()
```

```python
data.drop(['Gender_F','MaritalStatus_M'],axis=1,inplace=True)
data.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

gender_m = data['Gender_M'].values.reshape(-1, 1)
marital_status_s = data['MaritalStatus_S'].values.reshape(-1, 1)

similarity = cosine_similarity(gender_m, marital_status_s)
similarity
```

```python
intersection = len(set(data['Gender_M']).intersection(set(data['MaritalStatus_S'])))
union = len(set(data['Gender_M']).union(set(data['MaritalStatus_S'])))
if union == 0:
  print(0)
else:
  print( intersection / union)
```
